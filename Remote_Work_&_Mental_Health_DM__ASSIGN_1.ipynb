```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#loding my dataset in datafram 'remote_work'
remote_work = pd.read_csv('remote_work_mh.csv')

```

# pixel oriented visualization

```python
# For selecting only numeric data
data = remote_work.select_dtypes(include=[np.number])

# Normalize the numeric columns to a 0-1 range for visualization
normalized = (data - data.min()) / (data.max() - data.min())

# Converting the normalized dataframe to a 2D array for visualization
data_array = normalized.values

# Setting the plot size for presentation
plt.figure(figsize=(15, 10))

# Plotting the pixel-oriented visualization
plt.imshow(data_array, cmap='coolwarm', aspect='auto')

# Adding a color bar to represent the normalized value
cbar = plt.colorbar()
cbar.set_label('Normalized Feature Value', rotation=270, labelpad=20)

# Adding labels and title for better presentation
plt.title('Pixel-Oriented Visualization of Numeric Data', fontsize=16)
plt.xlabel('Features', fontsize=20)
plt.ylabel('Records', fontsize=20)

# Show the plot
plt.tight_layout()
plt.show()
```

# Expalnation

The code for pixel-oriented visualization starts by selecting only the numeric columns from the dataset to ensure accurate representation. It then normalizes this data to a 0-1 range, making all features comparable. The normalized data is converted into a 2D array suitable for plotting. The plot size is set for clarity, and the plt.imshow() function generates the visualization with a specified color scheme and aspect ratio. A color bar is added to indicate normalized values, and titles and labels enhance readability. Finally, the layout is adjusted for a clean appearance, and the completed plot is displayed.








# chernoff faces


```python
from matplotlib.patches import Ellipse, Circle   #its required fpr chernoff faces

# Function to draw Chernoff faces for each row of data
def draw_chernoff_face(ax, features):

    head = Ellipse((0.5, 0.5), width=1, height=1, facecolor='lightyellow')
    ax.add_patch(head)

    # For Eyes
    left_eye = Ellipse((0.3, 0.65), width=0.2, height=0.1 * features[0], facecolor='black')
    right_eye = Ellipse((0.7, 0.65), width=0.2, height=0.1 * features[1], facecolor='black')
    ax.add_patch(left_eye)
    ax.add_patch(right_eye)

    # For Mouth
    ax.plot([0.3, 0.7], [0.35, 0.35 + 0.1 * (features[2] - 0.5)], color='red', lw=2)

    # For Nose
    ax.plot([0.5, 0.5], [0.5, 0.45], color='black', lw=2)

    # For Eyebrows
    ax.plot([0.25, 0.35], [0.75, 0.75 + 0.1 * (features[3] - 0.5)], color='black', lw=2)
    ax.plot([0.65, 0.75], [0.75, 0.75 + 0.1 * (features[4] - 0.5)], color='black', lw=2)

remote_work= pd.read_csv('remote_work_mh.csv')

#using numeric features
numeric_df = remote_work.select_dtypes(include=[np.number])

# Normalizing the data to [0, 1] range
normalized_df = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

# Plotting Chernoff faces
num_faces = 9
rows = 3  # Number of rows for the grid
cols = 3  # Number of columns for the grid

fig, axs = plt.subplots(3, 3, figsize=(cols *3, rows *3))
for i, ax in enumerate(axs.flat):
    if i < num_faces:
        draw_chernoff_face(ax, normalized_df.iloc[i].values)
    draw_chernoff_face(ax, normalized_df.iloc[i].values)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

# Explanation

The code generates Chernoff faces to visualize data from the remote_work_mh.csv dataset, displaying up to 18 records in a 6x3 grid. It begins by loading the dataset and normalizing the numeric features to a range between 0 and 1. The draw_chernoff_face function creates faces based on these normalized values, adjusting elements like the eyes, mouth, nose, and eyebrows according to the feature values. The num_faces variable controls the number of faces shown, while the grid layout is defined by the rows and cols variables, enabling a clear representation of multiple data records.




# stick faces

```python
# Selecting numeric features
numeric_df = remote_work.select_dtypes(include=[np.number])

# Normalizing the data to [0, 1] range
normalized_df = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

# Function to draw stick faces for each row of data
def draw_stick_face(ax, features):
    # For head
    head = plt.Circle((0.5, 0.7), 0.1, color='black', fill=False)
    ax.add_patch(head)

    # For eyes
    left_eye = plt.Circle((0.45, 0.73), 0.02, color='black')
    right_eye = plt.Circle((0.55, 0.73), 0.02, color='black')
    ax.add_patch(left_eye)
    ax.add_patch(right_eye)

    # For mouth
    ax.plot([0.45, 0.55], [0.68, 0.68], color='black', lw=2)  # Straight mouth

    # For body
    ax.plot([0.5, 0.5], [0.6, 0.3], color='black', lw=2)  # Body

    # For arms
    arm_length = 0.1 * features[0]  # Adjust arm length based on the first feature
    ax.plot([0.5 - arm_length, 0.5 + arm_length], [0.5, 0.5], color='black', lw=2)

    # For legs
    leg_length = 0.1 * features[1]  # Adjust leg length based on the second feature
    ax.plot([0.5, 0.5 - leg_length], [0.3, 0.1], color='black', lw=2)
    ax.plot([0.5, 0.5 + leg_length], [0.3, 0.1], color='black', lw=2)

# Plotting stick faces for the first 9 rows of the dataset
num_faces = 25
rows = 5  # Number of rows for the grid
cols = 5  # Number of columns for the grid

fig, axs = plt.subplots(rows, cols, figsize=(cols * 5, rows * 5))
for i, ax in enumerate(axs.flat):
    if i < num_faces:
        draw_stick_face(ax, normalized_df.iloc[i].values)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

# Explanation

The code begins by loading a dataset containing 5001 records and 10 features, selecting only the numeric columns for visualization. It normalizes these features to a range between 0 and 1 to ensure consistent scaling. The draw_stick_face function is defined to create stick figures, where the head, eyes, mouth, body, arms, and legs are drawn using the normalized feature values to adjust the lengths of the arms and legs. Finally, stick faces are plotted for the first nine rows of data in a grid layout, making it easy to visualize the relationships between the different features in the dataset.




# Tree Map

```python
import squarify
# Creating DataFrame with the relevant columns i want to visualize
df = remote_work[['Work_Life_Balance_Rating', 'Social_Isolation_Rating', 'Company_Support_for_Remote_Work']]

# Calculating the mean values for the tree map
df_mean = df.mean()

# Using specific colors from the DataFrame
colors = ['#FF9999', '#66B3FF', '#99FF99']  # Customize your colors

# Setting the figure size
plt.figure(figsize=(10, 6))

# Creating the tree map
squarify.plot(sizes=df_mean, label=df_mean.index, color=colors, alpha=.8)

# Adding title and axis labels
plt.title('Tree Map of Impact of Remote Work on Mental Health', fontsize=16)
plt.axis('off')  # Hide the axes
plt.show()
```

# Explanation

This code snippet creates a tree map to visualize the mean values of three key factors related to remote work's impact on mental health: Work-Life Balance Rating, Social Isolation Rating, and Company Support for Remote Work. It starts by selecting the relevant columns from the dataset and calculating their mean values. Custom colors are assigned for each factor to enhance visual distinction. The tree map is generated using the squarify library, with specific sizes and labels based on the calculated means. Finally, the plot's title is added, and the axes are hidden for a cleaner presentation, effectively illustrating the overall impact of these factors on mental health in a visually appealing manner.
