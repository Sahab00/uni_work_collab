```python
import cv2
import numpy as np
from glob import glob

# ðŸŸ¢ Folder Path
dataset_path = "/content/5_dataset"

# ðŸŸ¢ Saari Images Load Karo
image_paths = glob(dataset_path + "/*.png")  # JPG images ke liye (PNG ho toh change karo)
X_data = []

for img_path in image_paths:
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Grayscale
    img = cv2.resize(img, (28, 28))  # Resize to 28x28
    img = img / 255.0  # Normalize (0-1)
    img = img.reshape(28, 28, 1)  # CNN ke liye reshape

    X_data.append(img)

# ðŸŸ¢ Convert to NumPy Array
X_data = np.array(X_data)

# ðŸŸ¢ Labels (Sab "5" hain, toh label 1 rakha)
y_data = np.ones(len(X_data))

print(f"\nâœ… {len(X_data)} images preprocess ho chuki hain!")
```
