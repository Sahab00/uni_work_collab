# LAB_TASK

```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("iris_data.csv")
```

# histogram

```python
# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(data['sepal_length'], kde=True)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

 Explanation: A histogram shows how data is distributed. It divides data into ranges (called "bins") and counts how many values fall into each range. This helps you understand the shape of the data, like whether it’s spread out, clustered, or follows a pattern (e.g., normal distribution).

# Bar Chart

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='species', data=data)
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
```

 Explanation:
 A bar chart is used to compare different categories. Each bar represents a category, and the height (or length) of the bar shows the value or count for that category. It’s helpful for visualizing comparisons between groups.

# 3. Line Chart

```python

# Simulating data over time
time_data = pd.DataFrame({
    'day': range(1, 11),
    'sepal_length': data['sepal_length'][:10]  # Assuming df is your original dataset
})

# Plotting the line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x='day', y='sepal_length', data=time_data)
plt.title('Line Chart of Sepal Length Over Time')
plt.xlabel('Day')
plt.ylabel('Sepal Length')
plt.show()
```

 Explanation::
 A line chart shows how data changes over time. The points are connected by lines, which makes it easy to see trends, such as increasing or decreasing values, over a period.

# Bar and Line Chart

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart for count of species
sns.countplot(x='species', data=data, ax=ax1, alpha=0.6, color='b')
ax1.set_ylabel('Count', color='b')
# Line chart for average sepal length of species
ax2 = ax1.twinx()
sns.pointplot(x='species', y='sepal_length', data=data, ax=ax2, color='r', markers='o', linestyles='-')
ax2.set_ylabel('Average Sepal Length', color='r')
plt.title('Bar and Line Chart of Species Count and Sepal Length')
plt.show()



```

 Explanation::
 This chart combines bars and a line on the same plot. The bars represent one set of data (like categories), while the line shows a trend over time or another variable. It's useful when you want to compare two related datasets.

# Pie Chart

```python
df = sns.load_dataset('iris')
# Pie Chart
species_count = data['species'].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Species Distribution')
plt.show()
```

 Explanation::
 A pie chart shows the parts of a whole. Each slice of the pie represents a category, and the size of the slice is proportional to the category's value. This is useful for showing how different parts contribute to a total.

# 6. Heatmap


```python
numeric_data = data.select_dtypes(include=[np.number])

# Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(numeric_data.corr(), cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

 Explanation:
A heatmap is used to show relationships between variables with colors. Darker or brighter colors represent higher or lower values. It’s great for finding patterns in large datasets, such as correlations between variables.

# 7.Pair Plot

```python
sns.pairplot(data)
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

 Explanation:
A pair plot shows relationships between pairs of variables in a dataset. It includes scatter plots for each pair and histograms on the diagonal. This is useful for understanding how different variables interact with each other.

# Box Plot

```python
# Box Plot
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```
